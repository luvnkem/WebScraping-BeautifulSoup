## Importing the relevant libraries

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
URL = 'https://books.toscrape.com/catalogue/page-'
```

## Loading the webpage

```python
# First we create an empty list to append our data to
books=[]
Price=[]
Stock_availability=[]
Books_url=[]
```

```python
# We create a For loop to get information from multiple items.
for page in range(1,5):
    response = requests.get(URL + str(page) + '.html')
    Soup = bs(response.text,'html.parser')
    print(Soup.prettify()) 
```

```python
# Grabbing the book titles
for article in Soup.find_all('article'):
    books.append(article.h3.a.text)
    print(books)
        
```

```python
#Grabbing the prices of each book
for article in Soup.find_all('article'):
    Price.append(article.find('div', class_ = 'product_price').p.text)
    print(Price)
```

```python
#Grabbing the stock availability of each book
for article in Soup.find_all('article'):
    Stock = Soup.find('p', class_ = 'instock availability').text.strip()
    Stock_availability.append(Stock)

print(Stock_availability)
```

```python
# We use a dictionary to pass in all the variables and store it in the object Scraped_Data 

Scraped_Data = {'TITLE':books,'PRICE':Price,'STOCK AVAILABILTY':Stock_availability}

Scraped_Data
```

```python
# We then load the dictionary in a pandas dataframe
Scraped_Books = pd.DataFrame(Scraped_Data)


Scraped_Books
```

```python
Scraped_Books.info()
```

```python
# Getting rid of the symbols at the begining of the object prices,

price = Scraped_Books['PRICE']
ACTUAL_PRICE = price.replace("^A","")
ACTUAL_PRICE
```

```python
# Saving the dataframe as a csv file with the name Scraped_file
Scraped_Books.to_csv('Scraped_file.csv',index = None)
```
